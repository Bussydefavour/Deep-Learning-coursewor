### Creating Plots for Each Column

For every numerical column in a given dataset, the code defines the function visualize_distribution_with_statistics, which produces boxplots, histograms, and statistical summary tables. The function displays the distribution of the data for each column as a boxplot, a histogram with a kernel density estimate, and a written summary of the most important statistics, including mean, median, standard deviation, and skewness. This all-inclusive method offers a thorough statistical and visual analysis of numerical data, assisting in the comprehension of data features like variability, central tendency, distribution shape, and outliers, all of which are essential for data exploration and preprocessing in statistical analysis and machine learning.

```python
# Creating plots for each column(boxplot, histogram, and a statistical summary table)
def visualize_distribution_with_statistics(dataset, numeric_columns):
    # Set up subplots with one row per column and three columns per row
    rows = len(numeric_columns)
    fig, axes = plt.subplots(rows, 3, figsize=(18, 4 * rows))

    # Flattening the axes for iteration
    axes = axes.flatten()

    # Defining color for plots
    boxplot_color = '#87CEEB'

    for i, column in enumerate(numeric_columns):
        # Creating boxplot for current column with title
        sns.boxplot(x=dataset[column], ax=axes[3*i], color=boxplot_color)
        axes[3*i].set_title(f'Boxplot of {column}')

        # Creating histogram for current column with kernel density estimation with title
        sns.histplot(dataset[column], kde=True, ax=axes[3*i+1], color=boxplot_color)
        axes[3*i+1].set_title(f'Histogram of {column}')

        # Adding text annotations for the mean, median, standard deviation, and skewness of the current column
        axes[3*i+2].text(0.05, 0.9, f'Mean: {dataset[column].mean():.2f}', transform=axes[3*i+2].transAxes)
        axes[3*i+2].text(0.05, 0.8, f'Median: {dataset[column].median():.2f}', transform=axes[3*i+2].transAxes)
        axes[3*i+2].text(0.05, 0.7, f'Std Dev: {dataset[column].std():.2f}', transform=axes[3*i+2].transAxes)
        axes[3*i+2].text(0.05, 0.6, f'Skewness: {dataset[column].skew():.2f}', transform=axes[3*i+2].transAxes)

    # Adjusting the spacing for subplots
    plt.tight_layout()
    # Displaying plots
    plt.show()

visualize_distribution_with_statistics(telcom_df, num_col)
```

### Dealing with Outliers

The code defines a function called IQR_method, which uses the Interquartile Range (IQR) method to find and list outliers in numerical columns of a dataset. It computes the IQR for each numerical column and classifies values that fall below Q1 minus 1.5 times the IQR or above Q3 plus 1.5 times the IQR as outliers. These outlier indices are then combined, counted, and returned by the function. The function is defined, then applied to the telcom_df DataFrame by the code, which finds outliers in the designated numerical columns (num_col). After identifying and eliminating these outliers, telcom_df is renamed telcom_df_out to reflect the absence of outliers. Outliers have the potential to greatly distort results, therefore this procedure is essential for cleaning data and guaranteeing the validity of statistical studies and machine learning models.

```python
def IQR_method(dataset, n, features):
    outlier_list = []

    for column in features:
        # Skipping categorical columns
        if pd.api.types.is_categorical_dtype(dataset[column]):
            continue

        # Calculate Q1 (25th percentile of the data) for the given feature
        Q1 = dataset[column].quantile(0.25)
        # Calculate Q3 (75th percentile of the data) for the given feature
        Q3 = dataset[column].quantile(0.75)
        # Use the interquartile range to calculate an outlier step (1.5 times the interquartile range)
        IQR = Q3 - Q1
        outlier_step = 1.5 * IQR

        # Determine a list of indices of outliers for the feature
        outlier_list_column = dataset[(dataset[column] < Q1 - outlier_step) | (dataset[column] > Q3 + outlier_step)].index
        outlier_list.extend(outlier_list_column)

    # Select observations containing more outliers
    outlier_list = dataset.loc[outlier_list]
    multiple_outliers = []

    # Calculate the number of records below and above the lower and upper bounds, respectively
    for column in features:
        if pd.api.types.is_categorical_dtype(dataset[column]):
            continue

        out1 = dataset[(dataset[column] < Q1 - outlier_step)]
        out2 = dataset[(dataset[column] > Q3 + outlier_step)]

        multiple_outliers.extend(out1.index)
        multiple_outliers.extend(out2.index)

    print('The total number of deleted outliers is:', len(multiple_outliers))

    return multiple_outliers

# Detecting and dropping outliers
Outliers_IQR = IQR_method(telcom_df, 1, num_col)
telcom_df_out = telcom_df.drop(Outliers_IQR, axis=0).reset_index(drop=True)
```

# Preprocess the Data for PyTorch

To prepare the telcom_df DataFrame for use in a PyTorch machine learning model, the code offers a step-by-step procedure. The target variable, "Churn," is first encoded using LabelEncoder. Next, categorical and numeric columns are identified, and make_column_transformer is used to apply one-hot encoding to categorical columns and standard scaling to numeric columns. After being converted, the data is reassembled into a DataFrame and divided into the target variable (y) and features (X). Train_test_split is used to further divide this data into training and testing sets. In order to train and test a PyTorch model, the feature and target sets are finally transformed into PyTorch tensors.

This preparation is essential for machine learning tasks ensuring that the data is in a format that PyTorch algorithms can use for model training and evaluation. It does this by standardising the input format, suitably scaling numeric data, and encoding categorical variables.

```python
# Identify categorical and numeric columns
cat_cols = telcom_df.select_dtypes(include=['object', 'category']).columns
num_cols = telcom_df.select_dtypes(include=['float64', 'int64']).columns

# Preprocess the data for PyTorch
def preprocess_data(telcom_df):
    # Temporarily store and remove 'Churn' column
    churn_col = telcom_df['Churn'].copy()
    df = telcom_df.drop('Churn', axis=1)

    # Encode target variable
    le = LabelEncoder()
    churn_encoded = le.fit_transform(churn_col)

    # OneHotEncode categorical data and StandardScaler for numeric data
    transformer = make_column_transformer(
        (OneHotEncoder(), cat_cols),
        (StandardScaler(), num_cols),
        remainder='passthrough'
    )

    # Apply transformations
    transformed = transformer.fit_transform(telcom_df)
    transformed_df = pd.DataFrame(transformed, columns=transformer.get_feature_names_out())

    # Add 'Churn' back to the DataFrame
    transformed_df['Churn'] = churn_encoded

    # Splitting into X and y
    X = transformed_df.drop('Churn', axis=1)
    y = transformed_df['Churn']

    return X, y

# Split data into training and testing sets
X, y = preprocess_data(telcom_df)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Convert the data into PyTorch tensors
X_train_tensor = torch.tensor(X_train.values, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test.values, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test.values, dtype=torch.float32)
```

## Defining the Neural Network Architecture

Using PyTorch's neural network module nn.Module, the code constructs a neural network architecture called churnPredictionModel for forecasting customer churn. There are two hidden layers and an output layer in the model. With a predetermined number of neurons, the first hidden layer (fc1) is fully linked, and the second hidden layer (fc2) has half as many neurons as the first. For binary classification, there is only one neuron in the output layer (output). The forward method specifies the network's forward pass, applying the sigmoid function to the output layer and the ReLU activation function to both hidden layers, making it appropriate for binary classification. The number of input features in the model is set to match the number of columns in X_train, and the number of neurons is predetermined.

```python
# Neural Network Architecture
class AdvancedChurnPredictionModel(nn.Module):
    def __init__(self, input_features):
        super(AdvancedChurnPredictionModel, self).__init__()
        self.fc1 = nn.Linear(input_features, 128)  # First hidden layer
        self.dropout1 = nn.Dropout(0.3)  # Dropout layer
        self.fc2 = nn.Linear(128, 64)  # Second hidden layer
        self.dropout2 = nn.Dropout(0.2)  # Another dropout layer
        self.output = nn.Linear(64, 1)  # Output layer

    def forward(self, x):
        x = F.relu(self.fc1(x))  # Activation function for first layer
        x = self.dropout1(x)  # Apply dropout
        x = F.relu(self.fc2(x))  # Activation function for second layer
        x = self.dropout2(x)  # Apply dropout
        x = torch.sigmoid(self.output(x))  # Sigmoid activation function for output
        return x

# Instantiate the model
input_features = X_train.shape[1]
model = AdvancedChurnPredictionModel(input_features)

# Move model to appropriate device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
X_train_tensor = X_train_tensor.to(device)
y_train_tensor = y_train_tensor.to(device)
X_test_tensor = X_test_tensor.to(device)
y_test_tensor = y_test_tensor.to(device)

# Loss Function and Optimizer
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

## Training and Evaluating the Model

In order to prepare a neural network model for customer churn prediction in PyTorch, the code first initialises an optimisation algorithm called Adam with a learning rate of 0.001 and a binary cross-entropy loss function (BCELoss). To manage the model training procedure, a training function called train_model is defined. The model, loss criterion, optimizer, training data, and epoch count are all accepted. The loss is computed and updated every ten epochs through the use of forward propagation and backpropagation inside each epoch. The constructed PyTorch tensors of the training dataset are then used to train this model. Furthermore, an evaluate_model function is developed to evaluate the accuracy of the model by calculating the accuracy by comparing the model's predictions on the test dataset with the actual labels.

```python
# Training Function
def train_model(model, criterion, optimizer, X_train, y_train, epochs=100):
    model.train()
    for epoch in range(epochs):
        optimizer.zero_grad()
        y_pred = model(X_train)
        loss = criterion(y_pred, y_train.view(-1, 1))
        loss.backward()
        optimizer.step()
        if epoch % 10 == 0:
            print(f'Epoch {epoch}, Loss: {loss.item()}')

# Evaluation Function
def evaluate_model(model, X_test, y_test):
    model.eval()
    with torch.no_grad():
        y_pred = model(X_test)
        correct = (y_pred.round() == y_test.view(-1, 1)).type(torch.float).sum().item()
        accuracy = correct / len(y_test)
    return accuracy

# Training and Evaluating the Model
train_model(model, criterion, optimizer, X_train_tensor, y_train_tensor, epochs=100)
accuracy = evaluate_model(model, X_test_tensor, y_test_tensor)
print(f'Accuracy: {accuracy}')
```

## Hyperparameter Tuning

This code iterates over various learning rates and neuron counts to perform hyperparameter tuning for the ChurnPredictionModel neural network. The model is initialised with the desired number of neurons for each combination, the Adam optimizer is configured with the desired learning rate, and the model is trained using the training set. It assesses the model's accuracy on the test set after training. The procedure records the optimal hyperparameters (learning rate and neuron count) and monitors the maximum accuracy. The code outputs the maximum accuracy attained and the matching ideal hyperparameters after evaluating every combination.

This approach is essential for optimizing the model's performance, ensuring that it is not only well-fitted to the training data but also generalizes effectively to unseen data, thereby improving the predictive power and robustness of the model.

```python
input_features = X_train.shape[1]  # Number of input features

best_accuracy = 0
best_params = {}

for lr in [0.001, 0.01, 0.1]:
    for neurons in [64, 128]:
        model = AdvancedChurnPredictionModel(input_features)
        optimizer = optim.Adam(model.parameters(), lr=lr)
        train_model(model, criterion, optimizer, X_train_tensor, y_train_tensor)
        accuracy = evaluate_model(model, X_test_tensor, y_test_tensor)
        if accuracy > best_accuracy:
            best_accuracy = accuracy
            best_params = {'learning_rate': lr, 'neurons': neurons}

print(f'Accuracy: {accuracy}')
print("Best Parameters:", best_params)
```
