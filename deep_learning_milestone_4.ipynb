# Preprocess the Data for PyTorch

Using scikit-learn's ColumnTransformer, the method particularly targets both numerical and categorical columns in order to preprocess the mile3_df dataset for machine learning in an organised manner. Let's dissect it:

Identifying Column Types: Mile3_df's columns are divided into numerical and category types in the first two lines. Columns are filtered using the select_dtypes method according to their data type. 'Object' and 'category' columns are commonly used for categorical data, and categorical_cols holds the names of all such columns. In numerical_cols, the names of columns that represent numerical data and are either 'float64' or 'int64' are stored.

**Setting up Preprocessing Steps**: Next, two primary transformers are used to produce the ColumnTransformer object, which is called preprocessor: ('num', numerical_cols, StandardScaler()): The numerical columns in this transformer are subjected to StandardScaler, which standardises characteristics by eliminating the mean and scaling to unit variance. ('cat', categorical_cols, OneHotEncoder()): In order to transform categorical data into one-hot numerical arrays, OneHotEncoder is applied to the category columns.

**Applying the Preprocessor**: Next, preprocessor.fit_transform(mile3_df) is used to fit and transform the preprocessor throughout the whole DataFrame. This operation encodes categorical data and standardises numerical data by applying the appropriate transformations to the corresponding columns.
The preprocessed data is contained in a NumPy array called processed_data, which is the outcome.

**Verifying the Result**: At last, the processed data's shape is printed. To ensure that the preprocessing was successful, this step is crucial. Because one-hot encoding spreads category features into numerous binary columns, processed_data will probably have more columns than mile3_df.

In a machine learning project, this procedure is essential because: (1) It guarantees the standardisation of numerical data, solving problems resulting from disparate scales. (2) It transforms categorical data into a format that machine learning algorithms can utilise efficiently. (3) By treating each column type correctly, ColumnTransformer enables clear, succinct, and maintainable code.

```python
# Identify categorical and numerical columns
categorical_cols = mile3_df.select_dtypes(include=['object', 'category']).columns
numerical_cols = mile3_df.select_dtypes(include=['float64', 'int64']).columns

# Preprocessing for categorical data (One-Hot Encoding) and numerical data (StandardScaler)
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(), categorical_cols)
    ])

# Applying the preprocessor to the dataset (Example with the entire dataset)
processed_data = preprocessor.fit_transform(mile3_df)

# Print the shape of the processed data to verify
print("Shape of processed data:", processed_data.shape)
```

## Splitting Dataset for Training, Evaluation and Testing

The mile3_df dataset can be thoroughly prepared for a machine learning model using the code below, especially for tasks like churn prediction.

**Preparing the Data for Splitting**: First, the dataset is split into the target variable (y) and features (X). The 'Churn' column—which is likely the target variable—is removed from the features by the drop method, but y is still given this column. Since the 'Churn' column is no longer included in the features (X), the lists of categorical and numerical columns have been changed to reflect this modification.

**Preprocessor Re-definition**: The ColumnTransformer preprocessor is redesigned using the new numerical and categorical columns. As before, this preprocessor will apply one-hot encoding to category columns and standardise numerical columns.

The dataset is divided into training, evaluation, and testing sets (train-test split). The test set (X_test, y_test), which makes up 10% of the original dataset, is first divided off using train_test_split. The remaining data is then split into training (i.e., 60% of the original dataset) and evaluation (30% of the original dataset) sets using train test split once more.

**Applying Transformations**: The training data (X_train) is processed by the preprocessor after it is fitted to the training data and transforms it.
The evaluation (X_eval) and test sets (X_test) are then transformed using the same preprocessor, which has already been fitted to the training set. To keep consistency, it's imperative to apply the same transformation settings to each data split.

**Verifying the Processed Data**: Lastly, printed out are the modified datasets' shapes. To ensure that the preprocessing stages have been applied appropriately, this step is crucial. The converted data will usually have more features than the original dataset because of one-hot encoding.

In machine learning, this procedure is essential to guarantee that: (1) The data is suitably divided into test, evaluation, and training sets for the purposes of training, fine-tuning, and assessing models. (2) The preprocessing methods for model consistency are the same for both evaluation and testing data (training and unseen). (3) To ensure that the changed data is ready for model training, it is checked.

```python
# Splitting the dataset into 60% training, 30% evaluation, and 10% testing
X = mile3_df.drop('Churn', axis=1)
y = mile3_df['Churn']

# Update categorical and numerical columns lists after dropping the target column
categorical_cols = X.select_dtypes(include=['object', 'category']).columns
numerical_cols = X.select_dtypes(include=['float64', 'int64']).columns

# Re-define the preprocessor with the updated lists
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
    ])

# Continue with the train-test split
X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
X_train, X_eval, y_train, y_eval = train_test_split(X_temp, y_temp, test_size=1/3, random_state=42)

# Apply transformations to the data
X_train = preprocessor.fit_transform(X_train)
X_eval = preprocessor.transform(X_eval)
X_test = preprocessor.transform(X_test)

# Print the transformed data shape to verify the preprocessing
print("Transformed training data shape:", X_train.shape)
print("Transformed evaluation data shape:", X_eval.shape)
print("Transformed testing data shape:", X_test.shape)
```

This code carries out the entire process of preparing data, training the model, and computing evaluation metrics for a neural network intended for binary classification (churn prediction) using PyTorch. Let's dissect it:

**Data Preprocessing**: The preprocessor, which was previously defined to handle both numerical and categorical data, is applied to the whole dataset (apart from the target column 'Churn'). ## Hyperparameter Tuning

This code uses PyTorch to offer a methodical approach to hyperparameter tuning for a neural network model. It looks at several combinations of batch sizes, learning rates, and epoch counts to see which are optimal for the churn prediction model. Let's examine the specifics:

Establishing Hyperparameter Ranges: lists with various values for each hyperparameter are used to create the learning rates, batch sizes, and num epochs choices. These are the options you should investigate to see which combination results in the optimal performance of the model.

train and assess the model Function: This function trains the ChurnPredictor model based on inputs such as learning_rate, batch_size, and num_epochs.
A new instance of the model is created inside the function, and then the optimizer (Adam) with the desired learning rate and the loss function (BCELoss) are defined.
The specified batch size is used to define data loaders for both training and evaluation sets. Throughout the predetermined number of epochs, the training loop modifies the model weights. Predictions are produced by the evaluation loop and utilised to determine the accuracy, F1 score, and R2 score. These evaluation metrics are returned by the function.

Iterating Through Combinations of Hyperparameters: To generate every conceivable combination of the values in learning_rates, batch_sizes, and num_epochs_options, utilise itertools.product. The train_and_evaluate_model function is run for each combination, and the performance metrics that are produced are printed.

Performance Metrics Output: The function outputs the learning rate, batch size, and number of epochs utilised together with the relevant performance metrics (accuracy and F1 score) after training and assessing the model with each hyperparameter set.

In machine learning, this hyperparameter tuning procedure is crucial since it finds a model's optimal parameters. Through methodical experimentation with various configurations, you may ascertain which hyperparameters yield optimal results based on your assessment metrics.

```python
import itertools

# Define a range of values for each hyperparameter
learning_rates = [0.001, 0.01]
batch_sizes = [32, 64]
num_epochs_options = [5, 10]

# This function trains the model and returns the evaluation metrics
def train_and_evaluate_model(learning_rate, batch_size, num_epochs):
    model = ChurnPredictor(input_size)
    criterion = nn.BCELoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)

    # DataLoader for batch processing
    train_loader = DataLoader(dataset=train_dataset, batch_size=batch_size, shuffle=True)
    eval_loader = DataLoader(dataset=eval_dataset, batch_size=batch_size, shuffle=False)

    # Training Loop
    for epoch in range(num_epochs):
        model.train()
        for inputs, labels in train_loader:
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

    # Evaluation Loop
    model.eval()
    y_pred, y_true = [], []
    with torch.no_grad():
        for inputs, labels in eval_loader:
            outputs = model(inputs)
            predicted = (outputs.data > 0.5)
            y_pred.extend(predicted.numpy())
            y_true.extend(labels.numpy())

    y_pred = [x[0] for x in y_pred]
    y_true = [x[0] for x in y_true]

    accuracy = accuracy_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)

    return accuracy, f1

# Iterate over all combinations of hyperparameters
for lr, batch_size, num_epochs in itertools.product(learning_rates, batch_sizes, num_epochs_options):
    accuracy, f1 = train_and_evaluate_model(lr, batch_size, num_epochs)
    print(f"LR: {lr}, Batch Size: {batch_size}, Epochs: {num_epochs}")
    print(f"Accuracy: {accuracy:.4f}, F1 Score: {f1:.4f}\n")
```

ROC (Receiver Operating Characteristic) and Precision-Recall curves are used in this code to visualise the performance of a trained neural network model (churn_predictor) for binary classification.

**assess_model Function:** The model is assessed using the provided dataset (eval_loader) using this function. The model's gradients are not computed in the evaluation mode (model.eval()), which uses less memory.

**Using the evaluate_model function**, the churn_predictor model is assessed, yielding output probabilities (y_probs) and true labels (y_true).

C**alculating Precision-Recall and ROC Curves**: sklearn provides the roc_curve and precision_recall_curve methods.The ROC and Precision-Recall curves are calculated, respectively, using metrics. comprehension the trade-offs between sensitivity (true positive rate), specificity (false positive rate), recall, and precision in classification tasks requires a comprehension of these curves.

**Plotting the ROC Curve**: The ROC curve is the focus of the first subplot.
The model's ability to discriminate between the classes is indicated by the area under the curve (AUC), which is plotted against the false positive rate (FPR) and the true positive rate (TPR). A model with a higher AUC is thought to perform better.
The Precision-Recall Curve Plotted:

The **Precision-Recall curve** is displayed in the second subplot. When working with unbalanced datasets, this curve is extremely instructive. It illustrates how recall—the ratio of true positives to all actual positives—and precision—the ratio of genuine positives to all anticipated positives—trade off.

Visualisation: To plot both curves with the proper labels, legends, and axis names for clarity, utilise Matplotlib.
The resulting plot is guaranteed to include no overlapping content thanks to plt.tight_layout().
Given that the visualisations are crucial for understanding the outcomes of binary classification models and making defensible decisions regarding their application, the utilisation of both ROC and Precision-Recall curves offers a thorough understanding of the model's performance by highlighting various facets of its predictive capabilities.
